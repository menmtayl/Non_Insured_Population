```python
import pandas as pd
df = pd.read_csv('the-number-of-estimated-eligible-uninsured-people-for-outreach-targeting.csv')
df.info()
print(df)
```

```python
df = df.drop_duplicates()
df
```

```python
df.info()
print(df)
```

```python
df['Foodstamp'].unique()
df[df['Foodstamp']==5]

df[df['Foodstamp'].isna()]
df['Foodstamp'].fillna(0, inplace=True)

df[df['Age0_18'].isna()]
df['Age0_18'].fillna(0, inplace=True)

df[df['Spanish'].isna()]
df['Spanish'].fillna(0, inplace=True)

df[df['Age26_34'].isna()]
df['Age26_34'].fillna(0, inplace=True)
```

```python
df["StateFIPS"] = df["StateFIPS"].astype(bool)
df.dtypes

df["PumaState"] = df["PumaState"].astype(bool)
df.dtypes

df["PUMA"] = df["PUMA"].astype(bool)
df.dtypes

df["StateName"] = df["StateName"].astype(bool)

df["Counties Within PUMA"] = df["Counties Within PUMA"].astype(bool)

```

```python
columns_to_convert = ['Foodstamp', 'Age0_18', 'Age26_34']
df[columns_to_convert] = df[columns_to_convert].astype(int)
print(df.dtypes)

```

```python
char_to_remove = '**', 'NaN', '...'
df = df.apply(lambda x: x.str.replace(char_to_remove,'')if x.dtype == 'object' else x)
print("\nDataFrame after removing character:")
print(df)
```
